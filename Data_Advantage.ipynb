```python
import itertools

from pokemon_env import *
from model import*
```

```python
env = PokemonEnv(num_battles=1, deterministic=False)
#env.build_pool(['Gastly', 'Jigglypuff'])
#state = env.reset()
#print(state)
```

```python
pokemon_list = list(itertools.combinations(list(env.pokedex), 2))
data = []
for pool in pokemon_list:
    print(pool)
    env.build_pool(pool)
    results = env.battle(render=False)
    data.append([pool[0], pool[1], results[1][0], results[1][1]])
```
