```python
import random
import itertools
import numpy as np
from tqdm import tqdm

from pokemon_env import *
from model import*
```

```python
env = PokemonEnv(num_battles=1, deterministic=False)
```

```python
pokemon_list = list(itertools.combinations(list(env.pokedex), 2))
data = []
for i in tqdm(range(1)):
    #random.shuffle(pokemon_list)
    for pool in pokemon_list:
        env.build_pool(pool)
        results = env.battle(render=False)
        winner_index = np.argmax(results[1]) 
        data.append([pool[0], pool[1], pool[winner_index]])
```

```python
df = pd.DataFrame(data, columns = ['Pokemon1', 'Pokemon2', 'Winner'])
df
```
